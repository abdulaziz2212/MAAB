3. **Nested Grouping on Flights**
   - Group flights by `Year` and `Month` and calculate:
     - Total number of flights.
     - Average arrival delay (`ArrDelay`).
     - Maximum departure delay (`DepDelay`).

```python
import pandas as pd

try:
    # Reading the data from parquet file
    flights = pd.read_parquet('flights')
    
    # Grouping by Year and Month
    grouped_flights = flights.groupby(['Year', 'Month']).agg(
        Total_Flights=('FlightNum', 'count'),
        Avg_ArrDelay=('ArrDelay', 'mean'),
        Max_DepDelay=('DepDelay', 'max')
    ).reset_index()

    # Displaying the results
    print(grouped_flights)

except FileNotFoundError:
    print("The file 'flights' was not found. Please check the file path and try again.")
except pd.errors.EmptyDataError:
    print("The file 'flights' is empty. Please provide a file with data.")
except Exception as e:
    print(f"An error occurred: {e}")
```

#### **Using `pipe`**
1. **Pipeline on Titanic**
   - Create a pipeline to:
     - Filter passengers who survived (`Survived == 1`).
     - Fill missing `Age` values with the mean.
     - Create a new column, `Fare_Per_Age`, by dividing `Fare` by `Age`.
     

```python
def survived(df):
    return df[df['Survived']==1]

df_survived = df_titanic.pipe(survived)
df_survived
```

```python
def fill_age(df):
    df['Age']=df['Age'].fillna(df['Age'].mean())
    return df

df_survived.pipe(fill_age)
```

```python
def new_column(df):
    df['Fare_Per_Age'] = df["Fare"] / df["Age"]
    return df

df_survived.pipe(new_column)
```

 **Pipeline on Flights**
   - Create a pipeline to:
     - Filter flights with a departure delay greater than 30 minutes.
     - Add a column `Delay_Per_Hour` by dividing the delay by the scheduled flight duration.

```python
import pandas as pd

# Load dataset from Parquet file
df_flights = pd.read_parquet('flights.parquet')

# Define pipeline functions
def filter_delayed_flights(df):
    """Keep only flights with a departure delay greater than 30 minutes."""
    return df[df['DEP_DELAY'] > 30]

def add_delay_per_hour(df):
    """Create a new column Delay_Per_Hour, handling zero durations to avoid division errors."""
    df = df.copy()  # Avoid modifying a slice of the DataFrame
    df['Delay_Per_Hour'] = df['DEP_DELAY'] / df['SCHEDULED_DURATION'].replace(0, pd.NA)
    return df

# Apply pipeline using pipe()
df_flights_processed = (df_flights
                        .pipe(filter_delayed_flights)  # Step 1: Filter delayed flights
                        .pipe(add_delay_per_hour)      # Step 2: Create Delay_Per_Hour column
                       )

# Display result
df_flights_processed.head()
```
