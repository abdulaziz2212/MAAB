```python
# Task 1: Union of Sets
def union_of_sets(set1, set2):
    return set1.union(set2)

# Task 2: Intersection of Sets
def intersection_of_sets(set1, set2):
    return set1.intersection(set2)

# Task 3: Difference of Sets
def difference_of_sets(set1, set2):
    return set1.difference(set2)

# Task 4: Check Subset
def is_subset(set1, set2):
    return set1.issubset(set2)

# Task 5: Check Element
def check_element_set(set1, element):
    return element in set1

# Task 6: Set Length
def set_length(set1):
    return len(set1)

# Task 7: Convert List to Set
def list_to_set(lst):
    return set(lst)

# Task 8: Remove Element
def remove_element_set(set1, element):
    set1.discard(element)
    return set1

# Task 9: Clear Set
def clear_set(set1):
    set1.clear()
    return set1

# Task 10: Check if Set is Empty
def is_empty_set(set1):
    return len(set1) == 0

# Task 11: Symmetric Difference
def symmetric_difference(set1, set2):
    return set1.symmetric_difference(set2)

# Task 12: Add Element
def add_element_set(set1, element):
    set1.add(element)
    return set1

# Task 13: Pop Element
def pop_element_set(set1):
    return set1.pop() if set1 else None

# Task 14: Find Maximum
def max_element_set(set1):
    return max(set1) if set1 else None

# Task 15: Find Minimum
def min_element_set(set1):
    return min(set1) if set1 else None

# Task 16: Filter Even Numbers
def filter_even_numbers_set(set1):
    return {x for x in set1 if x % 2 == 0}

# Task 17: Filter Odd Numbers
def filter_odd_numbers_set(set1):
    return {x for x in set1 if x % 2 != 0}

# Task 18: Create a Set of a Range
def create_range_set(start, end):
    return set(range(start, end + 1))

# Task 19: Merge and Deduplicate
def merge_and_deduplicate(lst1, lst2):
    return set(lst1).union(lst2)

# Task 20: Check Disjoint Sets
def are_disjoint_sets(set1, set2):
    return set1.isdisjoint(set2)

# Task 21: Remove Duplicates from a List
def remove_duplicates_list(lst):
    return list(set(lst))

# Task 22: Count Unique Elements
def count_unique_elements(lst):
    return len(set(lst))

# Task 23: Generate Random Set
def generate_random_set(size, start, end):
    import random
    return {random.randint(start, end) for _ in range(size)}
```
